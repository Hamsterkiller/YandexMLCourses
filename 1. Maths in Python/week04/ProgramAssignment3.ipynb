В качестве исследуемого закона распределения выберем хи-квадрат распределение. Данное распределение имеет один параметр - k (число степеней свободы). 
Теоретическое мат. ожидание данного распределения равно k.
Теоретическая дисперсия равна 2*k.

```python
import scipy.stats as sts
import numpy as np
import math
import matplotlib.pylab as plt
```

```python
# зададим распределение
k = 5
chisq_rv = sts.chi2(k)
```

```python
# задаем значения аргумента
x_values = np.linspace(0, 20, 100)
y_values = chisq_rv.cdf(x_values)
%matplotlib inline
plt.plot(x_values, y_values)
plt.title("chi2 cumulative distribution")
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

Выведем на график получившееся распределение:

```python
y_density = chisq_rv.pdf(x_values)
y_sample = chisq_rv.rvs(100)
plt.plot(x_values, y_density)
plt.hist(y_sample, 50, normed = True)
plt.title("chi2 prob. density function vs random sample from chi2")
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Оценим распределение выборочного среднего случайной величины при разных объемах выборок.

```python
# генерация набора размером n и вычисление выборочного среднего
def genSetAndMean(n):
    y_sample = chisq_rv.rvs(n)
    res = np.mean(y_sample)
    return res
```

Возьмем размер выборки n = 5. Рассчитаем также параметры аппроксимирующего нормального распределения.

```python
# генерация выборочных средних
means_5 = [genSetAndMean(5) for i in range(1, 1001)]
# расчет параметров аппроксимирующего нормального распределения
# мат. ожидание такого нормального распределения равно мат. ожиданию исх. случ. величины
mu = k
# дисперсия аппрокисирующего норм. распределения = Dисх/n. Dисх = 2*k
Dn = 2*k/5
# искомое нормальное распределение:
norm_rv = sts.norm(loc=mu, scale=np.sqrt(Dn))
x = np.linspace(0,12,1000)
y = norm_rv.pdf(x)
# гистрограмма распределения выборочных средних
plt.hist(means_5, 20, normed = True)
plt.plot(x, y)
plt.title("hist of 1000 means of samples of size 5")
plt.ylabel('$density$')
plt.xlabel('$x$')
plt.show()
print("Выборочное среднее выборочных средних равно ", np.mean(means_5))
print("Выборочная дисперсия выборочных средних равна ", np.mean((means_5-np.mean(means_5))**2))
```
